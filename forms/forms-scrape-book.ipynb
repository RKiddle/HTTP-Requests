```python
import requests

payload = {'firstname': 'Ryan', 'lastname': 'Mitchell'}
headers = {'User-Agent': 'Mozilla/5.0...'}

try:
    # Adding a 10-second timeout
    r = requests.post(
        'http://pythonscraping.com/pages/files/processing.php',
        data=payload,
        headers=headers,
        timeout=10 
    )
    print(r.text)
except requests.exceptions.Timeout:
    print("The server timed out. It's likely down or overloaded.")
except requests.exceptions.RequestException as e:
    print(f"A network error occurred: {e}")
```
